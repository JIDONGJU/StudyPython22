```python
re = response.text
soup = BeautifulSoup(re,"html.parser")
confCase_list = soup.find_all('confcase')
confCaseRate_list = soup.find_all('confcaserate')
createDt_list = soup.find_all('createdt')
criticalRate_list = soup.find_all('criticalrate')
death_list = soup.find_all('death')
deathate_list = soup.find_all('deathrate')
gubun_list = soup.find_all('gubun')

#<태그시작> ~ </태그끝>
#0번째에 있는 text를 가져와 (의미있는 정보인지 확인)
#  


gubun_list[0].text
```
