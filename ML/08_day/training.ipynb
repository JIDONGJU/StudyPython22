```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

titanic = pd.read_csv('./data/06_(문제)타이타닉_분류_데이터셋.csv')
print(titanic.info())
print(titanic)

'''
Survived  0이면 사망, 1이면 생존	
Pclass    티켓 등급 : 1, 2, 3	
Name      승객 이름 :성, 호칭. 이름 Surname  Title. Firstname	
Sex       성별      : male, female	
Age       나이	
SibSp     형제 자매, 배우자의 합	
Parch     부모 자식의 합	
Ticket    티켓 번호	
Fare      요금 
Cabin     객실 번호	
Embarked  선착장 : C(Cherbourg), Q(Queenstown), S(Southampton)
'''
```

```python
data = titanic[['survived', 'pclass', 'sex', 'age', 'sibsp', 'parch', 'fare', 'embarked', 'who', 'adult_male', 'alone']]
data
```

```python
data = data.dropna(axis=0)
print(data.isnull().sum())
print(data.info())
```

```python
for i in data.columns:
    print(i, data[i].unique())
    
# 'survived'   y
# 'pclass'     y
# 'sex'        y
# 'age'        구간 나누면 y 가능
# 'sibsp'      y
# 'parch'      y
# 'fare'       구간 나누면 y 가능
# 'embarked'   y
# 'who'        y
# 'adult_male' y
# 'alone'      y
```

```python
data_dum = pd.get_dummies(data)
data_dum
```

```python
data_dum['adult_male'] = data_dum['adult_male'].astype(str)
data_dum['alone'] = data_dum['alone'].astype(str)
data_dum.info()
```

```python
data_dum = pd.get_dummies(data_dum)
data_dum.columns
```

```python
titanic_input = data_dum[['pclass', 'age', 'sibsp', 'parch', 'fare', 'sex_female',
'sex_male', 'embarked_C', 'embarked_Q', 'embarked_S', 'who_child',
'who_man', 'who_woman', 'adult_male_False', 'adult_male_True',
'alone_False', 'alone_True']]
titanic_target = data_dum['survived']
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(titanic_input,
                                                                      titanic_target,
                                                                      random_state=42)

print(train_input.shape)
print(train_target.shape)
print(test_input.shape)
print(test_target.shape)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

ss.fit(train_input, train_target)

train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)

print(ss.mean_, ss.scale_)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

lr.fit(train_scaled, train_target)

print('로지스틱 회귀분류')
print('훈련',lr.score(train_scaled, train_target),'테스트',lr.score(test_scaled, test_target))
print('계수',lr.coef_, 'y절편',lr.intercept_)
print()
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()

dt.fit(train_scaled, train_target)

print('의사결정나무')
print('훈련',dt.score(train_scaled, train_target),'테스트',dt.score(test_scaled, test_target))
print()
```

```python
print(dt.feature_importances_)
```

```python
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestClassifier

# 랜덤포레스트 훈련모델
rfc = RandomForestClassifier()

# 훈련하기
rfc.fit(train_input, train_target)

# 훈련, 테스트 검증
print(rfc.score(train_input, train_target))
print(rfc.score(test_input, test_target))

# 교차검증 진행
# return_train_score : 검증결과 반환받기
scores = cross_validate(rfc, train_input, train_target, return_train_score=True, n_jobs= -1)

# 최종 훈련평가 결과 및 검증결과
print(scores)

# 최종 훈련모델의 평가점수(정확도)
print('train_score 평균 =', scores['train_score'].mean())
print('test_score 평균 =', scores['test_score'].mean())
```

```python
from sklearn.ensemble import ExtraTreesClassifier

etc = ExtraTreesClassifier(n_jobs=-1,
                           random_state=42)

etc.fit(train_input, train_target)

print(etc.score(train_input, train_target))
print(etc.score(test_input, test_target))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(random_state=42)

gb.fit(train_input, train_target)

print(gb.score(train_input, train_target))
print(gb.score(test_input, test_target))


scores = cross_validate(gb, train_input, train_target,
                        return_train_score= True, n_jobs= -1)

scores
print(scores['train_score'].mean(), scores['test_score'].mean())
```

```python
from sklearn.ensemble import HistGradientBoostingClassifier

hgb = HistGradientBoostingClassifier(random_state=42)

scores = cross_validate(hgb, train_input, train_target,
                        return_train_score= True, n_jobs= -1)

scores
print(scores['train_score'].mean(), scores['test_score'].mean())
```

```python
from xgboost import XGBClassifier

xxgb = XGBClassifier(tree_method= 'hist', random_state= 42,)

scores = cross_validate(xxgb, train_input, train_target,
                        return_train_score= True, n_jobs= -1)

scores
print(scores['train_score'].mean(), scores['test_score'].mean())
```

```python
from lightgbm import LGBMClassifier

lgb = LGBMClassifier(tree_method= 'hist', random_state= 42,)

scores = cross_validate(lgb, train_input, train_target,
                        return_train_score= True, n_jobs= -1)

scores
print(scores['train_score'].mean(), scores['test_score'].mean())
```
