### Paired Samples t-test
- 하나의 집단에서 보통 전(before)/후(after)의 관계를 확인할 때 사용
- 사용되는 함수 : ttest_rel(x, y)
    - x = before, y = after
- 검증
    - : p-value > 0.05 (전후 상관없음 = 차이없음)
    - : p-value <= 0.05 (전후 상관있음 = 차이있음)

#### exe-1) 야식을 먹기전과 먹은후의 몸무게 차이 검정
- 귀무가설 : 야식을 먹기전과 먹은후의 몸무게에 차이가 없다
- 대립가설 : 야식을 먹기전과 먹은후의 몸무게에 차이가 있다

```python
import numpy as np
from scipy import stats

# 야식 먹기 전 몸무게 랜덤 생성
beforeWeight = np.random.randint(low=40, high=90, size=20)
print('beforeWeight =', beforeWeight)

# 야식 먹은 후 몸무게 랜덤 생성
tmp_list = []
for weight in beforeWeight :
    # 0.99-0.02 ~ 0.99+0.02
    tmp_list.append(weight * np.random.normal(0.99, 0.02)) # (기준값, +-값, 갯수)

afterWeight = np.array(tmp_list)
print('afterWeight =', afterWeight)
print(" ")
# 전/후 검증하기
result = stats.ttest_rel(beforeWeight, afterWeight)
print('검정 결과 =', result)
print('검정 통계량 = %.3f, p-value = %.3f' %(result))
print(" ")
if result[1] > 0.05:
    print("<해석>")
    print("1. p-value =",result[1].round(3)," > 0.05 이므로, 귀무가설 채택")
    print("2. 귀무가설 채택 : 야식을 먹기 전과 먹은 후의 몸무게에 차이가 없다.")
elif result[1] <= 0.05:
    print("<해석>")
    print("1. p-value =",result[1].round(3)," > 0.05 이므로, 귀무가설 기각")
    print("2. 대립가설 채택 : 야식을 먹기 전과 먹은 후의 몸무게에 차이가 있다.")
```

#### exe-2) 특강 전/후 시험점수의 차이 검정
- 귀무가설 : 특강 전/후 시험점수의 차이가 없다
- 대립가설 : 특강 전/후 시험점수의 차이가 있다

```python
import random
import numpy as np
import scipy.stats as stats
from seaborn.distributions import displot
```

```python
# 데이터 생성하기
# 특강전 데이터 : 랜덤하게 80을 기준으로 +- 10, 100개 생성
np.random.seed(0)
beforeStudy = np.random.randint(low=70, high=90, size=100)
print('beforeStudy =', beforeStudy)

# 특강후 데이터 : 랜덤하게 77을 기준으로 +- 10, 100개 생성
np.random.seed(1)
afterStudy = np.random.randint(low=67, high=87, size=100)
print('afterStudy =', afterStudy)
print(" ")

# 정규성 검사
resultSh1 = stats.shapiro(beforeStudy)
print("정규성 검정 결과 = ", resultSh1)
print("검정 통계량 = %.3f, p-value = %.3f" %(resultSh1))
if resultSh1[1] > 0.05:
    print("<해석>")
    print("p-value = ",round(resultSh1[1],3)," > 0.05 이므로, 정규성을 만족한다")
elif resultSh1[1] <= 0.05:
    print("<해석>")
    print("p-value = ",round(resultSh1[1],3)," <= 0.05 이므로, 정규성을 불만족한다")
    
print(" ")
resultSh2 = stats.shapiro(afterStudy)
# 정규성 검정
print("정규성 검정 결과 = ", resultSh2)
print("검정 통계량 = %.3f, p-value = %.3f" %(resultSh2))
if resultSh2[1] > 0.05:
    print("<해석>")
    print("p-value = ",round(resultSh2[1],3)," > 0.05 이므로, 정규성을 만족한다")
elif resultSh2[1] <= 0.05:
    print("<해석>")
    print("p-value = ",round(resultSh2[1],3)," <= 0.05 이므로, 정규성을 불만족한다")

# 정규성 시각화
import seaborn as sns
import matplotlib.pyplot as plt

# 한글처리
plt.rc("font", family  ="NanumGothic")

# x : 데이터 집합
# stats.norm : 정규 분포 표시
sns.distplot(beforeStudy, fit=stats.norm, color='blue')
plt.show()
sns.distplot(afterStudy, fit=stats.norm, color='red')
plt.show()

# 전/후 검정
result = stats.ttest_rel(beforeWeight, afterWeight)
print('검정 결과 =', result)
print('검정 통계량 = %.3f, p-value = %.3f' %(result))
if result[1] > 0.05:
    print("<해석>")
    print("1. p-value =",result[1].round(3)," > 0.05 이므로, 귀무가설 채택")
    print("2. 귀무가설 채택 : 특강 전/후 시험점수의 차이가 없다.")
elif result[1] <= 0.05:
    print("<해석>")
    print("1. p-value =",result[1].round(3)," > 0.05 이므로, 귀무가설 기각")
    print("2. 대립가설 채택 : 특강 전/후 시험점수의 차이가 있다.")
    
# 해석
```

```python
# 등분산성 검정

# levene 방법

sLevene = stats.levene(beforeStudy, afterStudy)
print("등분산성 levene 검증결과  =", sLevene)
print("검정 통계량 = %.3f, p-value = %.3f" %(sLevene))
print("""(해석)
p-value = 0.057 > 0.05 이므로, 등분산성을 만족한다""")
print(" ")

# fligner 방법

sFligner = stats.fligner(beforeStudy, afterStudy)
print("등분산성 levene 검증결과  =", sFligner)
print("검정 통계량 = %.3f, p-value = %.3f" %(sFligner))
print("""(해석)
p-value = 0.063 > 0.05 이므로, 등분산성을 만족한다""")
print(" ")

# bartlett 방법

sBartlett = stats.bartlett(beforeStudy, afterStudy)
print("등분산성 levene 검증결과  =", sBartlett)
print("검정 통계량 = %.3f, p-value = %.3f" %(sBartlett))
print("""(해석)
p-value = 0.244 > 0.05 이므로, 등분산성을 만족한다""")
```

#### exe-3) 환자 9명의 복부 수술전/후 몸무게 변화에 대한 차이 검정
- 귀무가설 : 복부 수술 전/후 몸무게 변화가 없다.
- 대립가설 : 복부 수술 전/후 몸무게 변화가 있다.

```python
# 데이터 생성
np.random.seed(3)
x1 = np.random.randint(low=50, high=90, size=100)

np.random.seed(4)
x2 = np.random.randint(low=45, high=85, size=100)
print(" ")

# 정규성 검정
resultSh1 = stats.shapiro(x1)
print("정규성 검정 결과 = ", resultSh1)
print("검정 통계량 = %.3f, p-value = %.3f" %(resultSh1))
if resultSh1[1] > 0.05:
    print("<해석>")
    print("p-value = ",round(resultSh1[1],3)," > 0.05 이므로, 정규성을 만족한다")
elif resultSh1[1] <= 0.05:
    print("<해석>")
    print("p-value = ",round(resultSh1[1],3)," <= 0.05 이므로, 정규성을 불만족한다")

resultSh2 = stats.shapiro(x2)
print("정규성 검정 결과 = ", resultSh2)
print("검정 통계량 = %.3f, p-value = %.3f" %(resultSh2))
if resultSh2[1] > 0.05:
    print("<해석>")
    print("p-value = ",round(resultSh2[1],3)," > 0.05 이므로, 정규성을 만족한다")
elif resultSh2[1] <= 0.05:
    print("<해석>")
    print("p-value = ",round(resultSh2[1],3)," <= 0.05 이므로, 정규성을 불만족한다")

# 정규성 시각화
sns.distplot(x1, fit=stats.norm, color='blue')
plt.show()
sns.distplot(x2, fit=stats.norm, color='red')
plt.show()

# 전/후 검정
result = stats.ttest_rel(x1, x2)
print('검정 결과 =', result)
print('검정 통계량 = %.3f, p-value = %.3f' %(result))
if result[1] > 0.05:
    print("<해석>")
    print("1. p-value =",result[1].round(3)," > 0.05 이므로, 귀무가설 채택")
    print("2. 귀무가설 채택 : 복부 수술 전/후 몸무게 변화가 없다.")
elif result[1] <= 0.05:
    print("<해석>")
    print("1. p-value =",result[1].round(3)," > 0.05 이므로, 귀무가설 기각")
    print("2. 대립가설 채택 : 복부 수술 전/후 몸무게 변화가 있다.")

# 해석
```
