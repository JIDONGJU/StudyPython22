```python
import numpy as np
```

```python
# 1. NAND w1, w2, theta
# 2. AND w1, w2, theta
# 3. OR w1, w2, theta
# 함수 and 논리를 출력하는 함수
# 함수 nand 논리를 출력하는 함수
# 함수 or 논리를 출력하는 함수
# 함수 xor 논리를 출력하는 함수
```

```python
X = np.array([[0,1,0,1],[0,0,1,1]]).T
```

```python
def AND(X):
    W = np.array([[0.1],[0.1]])
    theta = 0.1111111111111
    return np.where(np.dot(X,W) > theta, 1, 0)

def OR(X):
    W = np.array([[0.1],[0.1]])
    theta = 0
    return np.where(np.dot(X,W) > theta, 1, 0)

def NAND(X):
    W = np.array([[0.1],[0.1]])
    theta = 0.11
    return np.where(np.dot(X,W) < theta, 1, 0)

def XOR(X):
    x = np.c_[NAND(X), OR(X)]
    return AND(x)
```

```python
# XOR(X)
def XOR(X):
    W1 = np.array([[0.1,0.1],[0.1,0.2]])
    W2 = np.array([[0.1],[0.1]])
    b1 = 0.1
    b2 = 0.1
    layer1 = np.where(np.dot(X, W1) > b1 ,1,0)
    layer2 = np.where(np.dot(layer1,W2) > b2, 1,0)
    return layer2
```

```python
print(AND(X))
print(OR(X))
print(NAND(X))
print(XOR(X))
```

```python
w1s = np.linspace(0.1,1,10)
w2s = np.linspace(0.1,1,10)
thetas = np.linspace(0.1,1,10)

# AND
for w1 in w1s:
    for w2 in w2s:
        for theta in thetas:
            result = np.where(X[:,0]*w1 + X[:,1]*w2 > theta, 1, 0)
            if np.sum(result == np.array([0,0,0,1])) == 4:
                print(w1,w2,theta)
```

```python
result == np.array([0,0,0,1])
```

```python
# OR
for w1 in w1s:
    for w2 in w2s:
        for theta in thetas:
            result = np.where(X[:,0]*w1 + X[:,1]*w2 >= theta, 1, 0)
            if np.sum(result == np.array([0,1,1,1])) == 4:
                print(w1,w2,theta)
```

```python
# NAND
for w1 in w1s:
    for w2 in w2s:
        for theta in thetas:
            result = np.where(X[:,0]*w1 + X[:,1]*w2 <= theta, 1, 0)
            if np.sum(result == np.array([1,1,1,0])) == 4:
                print(w1,w2,theta)
```

```python
# XOR
for w1 in w1s:
    for w2 in w2s:
        for theta in thetas:
            result = np.where(X[:,0]*w1 + X[:,1]*w2 == theta, 1, 0)
            if np.sum(result == np.array([0,1,1,0])) == 4:
                print(w1,w2,theta)
```
