```python
# 지도 라이브러리
import folium
# 위경도 데이터 읽어들이기
import json
```

```python
starbucks_map = folium.Map(
    # 최초에 보여줄 지도의 중앙점 좌표
    # [위도, 경도]
    location = [37.573050, 126.979189],

    # 지도 스타일 (openstreetmap : 일반적으로 사용되는 도시형 스타일 지도)
    #tiles = "openstreetmap",
    #tiles = "Stamen Terrain", # 산림 위주
    tiles = "cartodbpositron", # 하천, 도로위주

    # 최초 보여줄 지도의 스케일
    zoom_start = 50
)
starbucks_map
```

```python
# 데이터프레임에 있는 위도/경도의 값을 각각 추출
# 반복문 이용
for i in range(len(seoul_starbucks)):
    lat = seoul_starbucks.loc[i,'위도']
    lon = seoul_starbucks.loc[i,'경도']
    #print(lat, lon)

    # 지도에 표시할 마킹 모양 등 스타일 지정
    folium.CircleMarker(
        # 위치
        location= [lat, lon],
        # 채우기 색상
        fill_color = 'green',
        # 채우기 불투명도
        fill_opacity = 1,
        # 테두리 색상
        color = 'yellow',
        # 테두리 굵기
        weight = 1,
        # 전체 원의 지름
        radius = 3
    ).add_to(starbucks_map)

starbucks_map
```

```python
# 스타벅스 매장 타입별로 마커 표시하기

starbucks_map2 = folium.Map(
    # 최초에 보여줄 지도의 중앙점 좌표
    # [위도, 경도]
    location = [37.573050, 126.979189],

    # 지도 스타일 (openstreetmap : 일반적으로 사용되는 도시형 스타일 지도)
    tiles = "openstreetmap",

    # 최초 보여줄 지도의 스케일
    zoom_start = 11
)



for i in range(len(seoul_starbucks)):
    lat = seoul_starbucks.loc[i,'위도']
    lon = seoul_starbucks.loc[i,'경도']
    store_type = seoul_starbucks.loc[i, '매장타입']
    #print(lat, lon)
    # general : gray, 3
    # reserve : blue, 5
    # generalDT : red, 5
    
    fillcolor = ""
    size = 1

    if store_type == 'general':
        fillcolor = 'gray'
        size = 3
    elif store_type == 'reserve':
        fillcolor = 'blue'
        size = 5
    elif store_type == 'generalDT':
        fillcolor == 'red'
        size = 5

    


    # 지도에 표시할 마킹 모양 등 스타일 지정
    folium.CircleMarker(
        # 위치
        location= [lat, lon],
        # 채우기 색상
        fill_color = fillcolor,
        # 채우기 불투명도
        fill_opacity = 1,
        fill = True,
        # 테두리 색상
        color = fillcolor,
        # 테두리 굵기
        weight = 1,
        # 전체 원의 지름
        radius = size
    ).add_to(starbucks_map2)

starbucks_map2
```
