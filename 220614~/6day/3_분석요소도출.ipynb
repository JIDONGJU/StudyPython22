### 월별/연령대별 본인부담비용 시각화분석

```python
# 수진자연령, 요양개시일자, 심결본인부담금
df_month_age = df_20TABLE_orag_cp[['수진자연령','요양개시일자','심결본인부담금']]
df_month_age
```

```python
df_9under = df_month_age[df_month_age['수진자연령'] <= 9]
df_9under.reset_index(drop=True)
df_9under['연령대'] = '9세이하'
df_9under
```

```python
df_10 = df_month_age[(df_month_age['수진자연령'] > 9) & (df_month_age['수진자연령'] <= 19)]
df_10.reset_index(drop=True)
df_10['연령대'] = '10대'
df_10
```

```python
df_20 = df_month_age[(df_month_age['수진자연령'] >= 20) & (df_month_age['수진자연령'] < 30)]
df_20.reset_index(drop=True)
df_20['연령대'] = '20대'
df_20
```

```python
df_30 = df_month_age[(df_month_age['수진자연령'] >= 30) & (df_month_age['수진자연령'] < 40)]
df_30.reset_index(drop=True)
df_30['연령대'] = '30대'
df_40 = df_month_age[(df_month_age['수진자연령'] >= 40) & (df_month_age['수진자연령'] < 50)]
df_40.reset_index(drop=True)
df_40['연령대'] = '40대'
df_50 = df_month_age[(df_month_age['수진자연령'] >= 50) & (df_month_age['수진자연령'] < 60)]
df_50.reset_index(drop=True)
df_50['연령대'] = '50대'
df_60 = df_month_age[(df_month_age['수진자연령'] >= 60) & (df_month_age['수진자연령'] < 70)]
df_60.reset_index(drop=True)
df_60['연령대'] = '60대'
df_70 = df_month_age[(df_month_age['수진자연령'] >= 70) & (df_month_age['수진자연령'] < 80)]
df_70.reset_index(drop=True)
df_70['연령대'] = '70대'
df_80 = df_month_age[(df_month_age['수진자연령'] >= 80) & (df_month_age['수진자연령'] < 90)]
df_80.reset_index(drop=True)
df_80['연령대'] = '80대'
df_90 = df_month_age[(df_month_age['수진자연령'] >= 90) & (df_month_age['수진자연령'] < 100)]
df_90.reset_index(drop=True)
df_90['연령대'] = '90대'
df_90
```

```python
df_month_age_any = pd.concat([df_9under,df_10,df_20,df_30,df_40,df_50,df_60,df_70,df_80,df_90], axis=0, sort=True)
df_month_age_any.info()
```

```python
# 요양개시일자에서 월 추출하기
# int64를 str로 변환
df_month_age_any['요양개시일자'] = df_month_age_any['요양개시일자'].astype(str)
df_month_age_any.info()
```

```python
df_month_age_any['요양개시월'] = df_month_age_any['요양개시일자'].str[4:6]
df_month_age_any
```

```python
df_month_age_any.to_excel('./files_new2/df_month_age_any.xlsx', index=False)
```

```python
# 모든 연령대를 1개의 데이터프레임으로 통합하기
df_month_age_any = pd.DataFrame()

df_month_age_any = pd.concat([df_month_age_any, df_9under], ignore_index=True)
df_month_age_any = pd.concat([df_month_age_any, df_10], ignore_index=True)
df_month_age_any = pd.concat([df_month_age_any, df_20], ignore_index=True)
df_month_age_any = pd.concat([df_month_age_any, df_30], ignore_index=True)
df_month_age_any = pd.concat([df_month_age_any, df_40], ignore_index=True)
df_month_age_any = pd.concat([df_month_age_any, df_50], ignore_index=True)
df_month_age_any = pd.concat([df_month_age_any, df_60], ignore_index=True)
df_month_age_any = pd.concat([df_month_age_any, df_70], ignore_index=True)
df_month_age_any = pd.concat([df_month_age_any, df_80], ignore_index=True)
df_month_age_any = pd.concat([df_month_age_any, df_90], ignore_index=True)

df_month_age_any = df_month_age_any.reset_index(drop=True)
print(len(df_month_age_any))
df_month_age_any
```
