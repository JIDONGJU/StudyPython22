- 딸기가격과 환경과 관련이 있는가?

```python
import scipy.stats as stats

corr = stats.pearsonr(total_dna['가격'],total_dna['습도'])
print('습도 상관계수 =',corr[0],'유의수준 =',corr[1])
corr = stats.pearsonr(total_dna['가격'],total_dna['강수'])
print('강수 상관계수 =',corr[0],'유의수준 =',corr[1])
corr = stats.pearsonr(total_dna['가격'],total_dna['온도'])
print('온도 상관계수 =',corr[0],'유의수준 =',corr[1])
corr = stats.pearsonr(total_dna['가격'],total_dna['바람'])
print('바람 상관계수 =',corr[0],'유의수준 =',corr[1])
```

```python
ctab = pd.crosstab(index=total_dna['가격'], columns=total_dna['습도'])
print(ctab)
print()
```
