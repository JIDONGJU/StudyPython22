```python
import pandas as pd
```

```python
file1 = './smart_farm.csv'
smart = pd.read_csv(file1)
smart
file2 = './normal_farm.csv'
normal = pd.read_csv(file2)
normal

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

#### 정규성 시각화

```python
col_name = ['초장','엽장','엽폭','엽병장','엽수','관부직경','소화수','착과수']
plt.rc("font", family = "NanumGothic")
plt.rcParams['figure.figsize'] = (10, 10)

for i in col_name:
    sns.distplot(smart[i],fit=stats.norm ,color='red')
    sns.distplot(normal[i],color='blue')
    plt.ylabel('count')
    plt.legend(['smart','standard','normal'],loc=1)
    plt.show()
```

```python
smart_list = list()
smart_list.append(round(smart['초장'].mean(),2))
smart_list.append(round(smart['엽장'].mean(),2))
smart_list.append(round(smart['엽폭'].mean(),2))
smart_list.append(round(smart['엽병장'].mean(),2))
smart_list.append(round(smart['엽수'].mean(),2))
smart_list.append(round(smart['관부직경'].mean(),2))
smart_list.append(round(smart['화방 꽃수(소화수)'].mean(),2))
smart_list.append(round(smart['착과수'].mean(),2))
smart_list
```

```python
smart_mean = pd.DataFrame(smart_list,
                          index = ['초장','엽장','엽폭','엽병장','엽수','관부직경','소화수','착과수'])
smart_mean.columns = ['스마트팜']
smart_mean
```

```python
normal_list = list()
normal_list.append(round(normal['초장'].mean(),2))
normal_list.append(round(normal['엽장'].mean(),2))
normal_list.append(round(normal['엽폭'].mean(),2))
normal_list.append(round(normal['엽병장'].mean(),2))
normal_list.append(round(normal['엽수'].mean(),2))
normal_list.append(round(normal['관부직경'].mean(),2))
normal_list.append(round(normal['소화수'].mean(),2))
normal_list.append(round(normal['착과수'].mean(),2))
normal_list
```

```python
normal_mean = pd.DataFrame(normal_list,
                           index = ['초장','엽장','엽폭','엽병장','엽수','관부직경','소화수','착과수'])
normal_mean.columns = ['일반농작']
normal_mean
```

```python
farm_mean = pd.merge(smart_mean, normal_mean, left_index=True, right_index=True, how='left')
farm_mean
```

```python
name1 = ['스마트팜']
smart.index = name1*35
smart
name2 = ['일반농작']
normal.index = name2*35
normal
```

```python
print(smart.info())
print(normal.info())
```

```python
smart.rename(columns={'화방 꽃수(소화수)' : '소화수'}, inplace=True)
smart = smart[['초장','엽장','엽폭','엽병장','엽수','관부직경','소화수','착과수']]
smart
```

```python
total = pd.concat([smart, normal], axis=0)
total
```

```python
total = total.reset_index()
total
```

#### 평균비교

```python
gmdf = total.groupby('index').mean().reset_index()
gmdf
```

```python
gmdf.rename(columns={'index' : '농작방법'}, inplace=True)
gmdf
```

```python
plt.rcParams['figure.figsize'] = (10, 10)

for i in col_name:
    plt.rc("font", family = "NanumGothic")
    sns.barplot(data=gmdf,
                x='농작방법',
                y='{}'.format(i))
    plt.show()
```
