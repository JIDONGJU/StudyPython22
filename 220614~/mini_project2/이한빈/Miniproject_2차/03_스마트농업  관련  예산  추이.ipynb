```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/malgun.ttf"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
```

```python
file_path = "./data/01_스마트팜 예산.csv"
df = pd.read_csv(file_path, encoding="EUC-KR")
df
```

```python
df
```

```python
df1 = df[['구분', '2014', '2015', '2016', '2017', '2018', '2019',
       '2020', '2021', '2022']].loc[:0]
df1
```

```python
df2 = df1.set_index("구분")
```

```python
df3 = df2.T
```

```python
df3["스마트농업 예산(A)"] = df3["스마트농업 예산(A)"].str.replace(",", "")
```

```python
df4 = df3.T.astype(int)
df4
```

```python
plt.rcParams['figure.figsize'] = (30,15)
plt.rcParams['lines.linewidth'] = 2
plt.rc("font", family = "Malgun Gothic", size=10)
df4.T.plot(kind="bar", fontsize=20)

plt.title("스마트팜 농업 예산", fontsize=40)
plt.xlabel(" ", fontsize=20)
a = range(0,20,1)

plt.xticks(fontsize = 20, rotation=0)
plt.yticks(fontsize = 20)

plt.legend(labels=["단위 : 억원"],loc="upper right", fontsize=20)
plt.grid(True)

plt.show()
```
