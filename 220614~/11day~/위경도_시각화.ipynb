## 위경도 시각화

### 부산광역시 전기차 충전소 위경도 시각화

```python
import pandas as pd
```

```python
file_path = './data/부산충전소.csv'
busan_charge_map = pd.read_csv(file_path, encoding='euc-kr')
busan_charge_map['충전기타입'].value_counts()
```

```python
import openpyxl
# 지도 라이브러리
import folium
# 위경도 데이터 읽어들이기
import json
```

```python
busan_map = folium.Map(
    # 최초에 보여줄 지도의 중앙점 좌표
    # [위도, 경도]
    location = [37.573050, 126.979189],

    # 지도 스타일 (openstreetmap : 일반적으로 사용되는 도시형 스타일 지도)
    #tiles = "openstreetmap",
    #tiles = "Stamen Terrain", # 산림 위주
    tiles = "cartodbpositron", # 하천, 도로위주

    # 최초 보여줄 지도의 스케일
    zoom_start = 50
)
busan_map
```

```python
# 스타벅스 매장 타입별로 마커 표시하기
default_map = folium.Map(
    # 초기 지도의 중앙점 좌표[위도, 경도]
    location=[35.150444, 129.059338],

    # 지도 스타일
    tiles='openstreetmap',
    # 초기 지도의 스케일
    zoom_start= 11
)

for i in range(len(busan_charge_map)):
    lat = busan_charge_map['lat'].iloc[i]
    lng = busan_charge_map['lng'].iloc[i]
    charge_type = busan_charge_map.loc[i, '충전기타입']

    fillcolor = ""
    size = 1
    
    if charge_type == 'AC완속':
        fillcolor = 'green'
        size = 3
    elif charge_type == 'DC콤보':
        fillcolor = 'blue'
        size = 5
    elif charge_type == 'DC차데모+AC3상+DC콤보':
        fillcolor == 'red'
        size = 5
    elif charge_type == 'DC차데모+DC콤보':
        fillcolor = 'white'
        size = 5
    elif charge_type == 'DC차데모':
        fillcolor == 'black'
        size = 5
    elif charge_type == 'DC차데모+AC3상':
        fillcolor = 'yellow'
        size = 5
    
    # 지도에 표시할 마킹 모양 등 스타일 지정
    folium.CircleMarker(
        location= [lat, lng],
        # 마커 채우기 색상
        fill_color = fillcolor,
        # 마커 채우기 불투명도
        fill_opacity = 1,
        fill=True,
        # 마커 테두리 색상
        color = fillcolor,
        # 마커 테두리 굵기
        weight =1,
        # 마커 지름
        radius= 5
    ).add_to(default_map)

default_map
```
