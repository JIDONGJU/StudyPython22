```python
import pandas as pd
# csv파일 불러오기
csv = pd.read_csv('./data/부산충전소.csv', encoding='euc-kr')
csv.head(10)
```

```python
import re

pattern = r'\([^)]*\)'
text = []

for i in range(0, len(csv['주소']),1):
    text.append(re.sub(pattern=pattern, repl='', string= csv['주소'][i]))
    

text
csv['주소'] = text
csv.head(10)
```

```python
# 가입 없이 주소->좌표 변환
from geopy.geocoders import Nominatim

def geocoding(address):
    geolocoder = Nominatim(user_agent = 'South Korea', timeout=None)
    geo = geolocoder.geocode(address)
    crd = {"lat": str(geo.latitude), "lng": str(geo.longitude)}

    return crd
```
