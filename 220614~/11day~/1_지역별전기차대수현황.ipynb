## 지역별 전기차 대수 현황

```python
import pandas as pd
```

```python
# 파일 불러오기
file = './data/지역별_전기차_대수_현황.csv'
df = pd.read_csv(file, encoding='euc-kr')
# 기준일로 정렬하기
df = df.sort_values(by='기준일').reset_index()
# 필요없는 컬럼 지우기
df = df.drop('index', axis=1)
df = df.drop('합계', axis=1)
print(len(df))
df
```

```python
# 데이터 추출 연습
df.loc[[0]][['서울','인천','경기','강원','충북','충남','대전','세종','경북','대구','전북','전남','광주','경남','부산','울산','제주']]
```

## 3년간 전기차 증가대수

```python
# 복사본 만들기
df1 = df.copy()
# 기준일 기준 마지막행 - 첫행
df1_1 = df1.loc[[0]][['서울','인천','경기','강원','충북','충남','대전','세종','경북','대구','전북','전남','광주','경남','부산','울산','제주']]
df1_2 = df1.loc[[37]][['서울','인천','경기','강원','충북','충남','대전','세종','경북','대구','전북','전남','광주','경남','부산','울산','제주']]
df1_2 = df1_2.reset_index().drop('index', axis=1)
# 증가율 만들기
df1_3 = (df1_2 / df1_1) * 100
df1_3
df1_4 = df1_2 - df1_1
```

```python
# 쉐잎 바꾸기
df1_3=df1_3.transpose()
df1_3
```
