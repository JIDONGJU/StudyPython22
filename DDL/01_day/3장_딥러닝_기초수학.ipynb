### 딥러닝을 위한 기초 수학
- 딥러닝을 배운다 라는 말은 딥러닝 실행법 뿐만 아니라 딥러닝의 수학원리를 공부하는 의미도 내포
- 원리를 알아야 정확히 실행할 수 있음
    - 어떤 원리로 입력 값의 패턴을 분석하고 학습하는지 > 수학연산, 함수
- 고등학교 수준의 수학만으로도 원리와 배경 파악 가능(더 나아가도 대학 교양 수준)
- 쫄지말자

### 일차 함수, 기울기와 Y절편
- 함수 : 두 집합 사이의 관계를 설명하는 수학 개념
    - 변수 x와 y가 있을 때, x가 변하면 y는 어떤 규칙으로 변하는지 나타냄
    - y = f(x) 로 표현
- 일차 함수 : y가 x에 관한 일차식으로 표현된 경우를 의미
    - = y = ax + b (a는 0이 아님)
    - a는 기울기, b는 절편

### 이차 함수의 최솟값
- 이차 함수 : y가 x에 관한 이차식으로 표현되는 경우
    - y = ax^2 (a는 0이 아님)
    - a > 0이면 아래로 볼록한 그래프, a < 0이면 위로 볼록한 그래프
    - 
    - y = ax^2 그래프를 x축 방향으로  p만큼, y축 방향으로 q만큼 평행 이동 시키면
    - y = a(x-p)^2 + q 로 표현
    - 포물선의 맨 아래에 위치한 지점을 최솟값
    - 
    - 이 최솟값은 '최소 제곱법'으로 쉽게 알아내는데 딥러닝에서 써먹기 힘듬
    - 최소 제곱법을 계산하기 위해 필요한 조건들을 알 수 없기 때문에
    - 따라서 '미분'과 '기울기'를 이용

### 미분, 순간 변화율과 기울기
- 딥러닝을 이해하는데 가장 중요한 수학은 '미분'
- 딥러닝은 결국 일차 함수의 a와 b 값을 구하는 것인데, a와 b 값은 이차 함수 포물선의 최솟값을 구하는 것
- 
- y = ax^2 그래프가 있다고 가정
- 오른쪽이나 왼쪽으로 '0에 가까울만큼' 이동한다고 가정
- 실제로 움직이는 것이 아니라 방향만 드러내는 정도의 순간적인 변화만 있을 것
- 이 순간의 변화를 '순간 변화율' 이라 함
- 이 방향을 따라 직선을 그려주면 '기울기'
- 
- 즉, 미분을 한다는 것은 '순간 변화율'을 구하는 것
- 어느 순간에 어떤 변화가 일어나고 있는지 숫자로 나타낸 것이 '미분계수' = 기울기
- 기울기가 0일때, 그래프에서 최솟값인 지점
- 기울기는 y값 증가량 / x값 증가량
- 
- 순간 변화율은 극한(limit) 기호를 사용

### 미분의 기본 공식
1. f(x) = x일 때 f '(x) = 1
2. f(x) = a에서 a가 상수일 때 f '(x) = 0
3. f(x) = ax에서 a가 상수일 때 f '(x) = a
4. f(x) = xa에서 a가 자연수일 때 f '(x) = axa-1
5. f(g(x))에서 f(x)와 g(x)가 미분 가능할 때 {f(g(x))}' = f '(g(x)) × g'(x)

### 편미분
- 원하는 한 가지 변수만 미분하고 그 외에는 모두 상수로 취급하는 것
- f(x,y) = x^2 + yx + a 일때
    - f'(x,y) = 2x + y

### 지수와 지수함수
- a^ㅁ 의 형태로 a를 '밑', ㅁ를 '지수'라고 함, a를 ㅁ만큼 곱한다는 뜻
- '지수함수'란 변수 x가 지수 자리에 있는 경우를 의미
- y = a^x (a는 1이 아니며 0보다 큼)

### 시그모이드 함수
- 딥러닝의 내부를 보면 입력받은 신호를 얼마나 출력할지 게산하는 과정이 무수히 반복
- 이때 출력 값으로 얼마나 내보낼지 계산하는 함수를 '활성화 함수'라고 함
- 활성화 함수로 가장 먼저 배우는 '시그모이드 함수'
- 
- 시그모이드 함수는 지수 함수에서 밑 값이 자연 상수인 함수를 의미
- f(x) = 1 / 1 + e^-x
- 
- 시그모이드 함수를 그래프로 그리면 S자 형태
- x가 큰 값을 가지면 f(x)는 1에 가까워지고, 작은 값을 가지면 0에 가까워짐
- 이 함수의 속성은 0 또는 1로 하나를 고를 때 유용하게 쓰임

### 로그와 로그함수
- 로그를 이해하려면 지수부터 이해해야 함
- a^x = b 일때, a와 b를 알고 있는데 x를 모를때 x를 구하기 위해 사용하는 방법이 로그
- 영어로 Logarithm 이라고 함
- 로그함수와 지수함수는 역함수의 관계

```python
fig = plt.figure(figsize=(12,12))
ax0 = fig.add_subplot(1,3,1)
ax1 = fig.add_subplot(1,3,2)
ax2 = fig.add_subplot(1,3,3)

ax0.imshow(Image.open('./data/ex/log.jpg'))
ax1.imshow(Image.open('./data/ex/log2.jpg'))
ax2.imshow(Image.open('./data/ex/log3.jpg'))
```
